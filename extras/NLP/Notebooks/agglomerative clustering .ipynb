```python
from sklearn.cluster import AgglomerativeClustering
```

```python
corpus = open('dataset.txt').read()
```

```python
docs = corpus.split('\n')
```

```python
x =[]
for doc in docs:
    i
    x.append(i)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec_obj = CountVectorizer()
```

```python
struct_data = vec_obj.fit_transform(x)
```

```python
struct_data.toarray()
```

```python
aggc = AgglomerativeClustering()
```

```python
aggc.fit(struct_data.toarray()) # for clustring data must be in the array else got error
```

```python
aggc.labels_
```
